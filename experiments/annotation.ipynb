```python
categorias = {}
with open('data/categorias.txt', 'r') as file_categorias:
    print('Lendo Categorias...', end='\n\n')
    for line in file_categorias.readlines():
        line = line.split(':')
        categoria = line[0]
        termos = line[1:]
        for termo in termos:
            if termo.endswith('\n'):
                termo = termo[:-1]
            if categoria in categorias.keys():
                categorias[categoria] += str(termo)
            else:
                categorias[categoria] = str(termo)
```

```python
categorias
```

```python
regexes = {'homofobico': '', 'obsceno': '', 'racismo': '',
           'insulto':'', 'misogino': '', 'xenofobia': ''}

for categoria, termos in categorias.items():
    for termo in termos.split(','):
        regexes[categoria] += str(termo) + '|'
# Removendo o ultimo '|'
for categoria in regexes.keys():
    regexes[categoria] = regexes[categoria][:-1]
```

```python
twitter_data['toxico'] = np.zeros((twitter_data.shape[0]))
twitter_data['homofobico'] = np.zeros((twitter_data.shape[0]))
twitter_data['obsceno'] = np.zeros((twitter_data.shape[0]))
twitter_data['racismo'] = np.zeros((twitter_data.shape[0]))
twitter_data['insulto'] = np.zeros((twitter_data.shape[0]))
twitter_data['misogino'] = np.zeros((twitter_data.shape[0]))
twitter_data['xenofobia'] = np.zeros((twitter_data.shape[0]))
```

```python
twitter_data_sample = twitter_data.sample(1000, random_state=42)
```

```python
for i, tweet in twitter_data_sample.iterrows():
    for category, regex in regexes.items():
        if re.search(regex, tweet['tweet_text'].lower()):
            twitter_data_sample.loc[i, 'toxico'] = 1
```

```python
twitter_data_sample[twitter_data_sample['toxico'] == 0].sample(30, random_state=0).to_csv('data/anotacao/negative_twitter.csv')
```
